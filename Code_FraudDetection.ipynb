```python
def check_duplicates(df):
    # Check if there are any duplicate rows in the DataFrame
    duplicates = df.duplicated()

    # Return True if duplicates exist, False otherwise
    return duplicates.any()

# Test the function
csv_file = "creditcard.csv"
df = pd.read_csv(csv_file)
result = check_duplicates(df)
print(result)
```

```python
def get_duplicate_rows(df):
    # Read the dataset into a Pandas DataFrame
    df = pd.read_csv(df)

    # Find duplicate rows in the DataFrame
    duplicate_mask = df.duplicated()

    # Use boolean indexing to get the duplicate rows
    duplicate_rows = df[duplicate_mask]

    return duplicate_rows

# Test the function
csv_file = "creditcard.csv"
duplicate_rows = get_duplicate_rows(csv_file)
print(duplicate_rows)
```

```python
def get_total_duplicate_rows(df):
    # Read the dataset into a Pandas DataFrame
    df = pd.read_csv(df)

    # Find duplicate rows in the DataFrame
    duplicate_mask = df.duplicated()

    # Get the total number of duplicated rows
    total_duplicates = duplicate_mask.sum()

    return total_duplicates

# Test the function
csv_file = "creditcard.csv"
total_duplicates = get_total_duplicate_rows(csv_file)
print("Total duplicated rows:", total_duplicates)
```

```python
def remove_duplicates(df):
    # Read the dataset into a Pandas DataFrame
    df = pd.read_csv(df)

    # Remove duplicate rows and keep only the unique rows
    df_unique = df.drop_duplicates()

    return df_unique

# Test the function
csv_file = "creditcard.csv"
unique_rows_df = remove_duplicates(csv_file)
print(unique_rows_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

def remove_duplicates(df):
    # Read the dataset into a Pandas DataFrame
    df = pd.read_csv(df)

    # Remove duplicate rows and keep only the unique rows
    df_unique = df.drop_duplicates()

    return df_unique

# Test the function
csv_file = "creditcard.csv"
unique_rows_df = remove_duplicates(csv_file)

# Suppress warnings related to deprecated distplot
with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    
    # Plot the distribution of 'Time' feature using distplot
    plt.figure(figsize=(18, 4))
    sns.distplot(unique_rows_df['Time'].values / (60 * 60), color='r')

    # Plot the distribution of 'Amount' feature using distplot
    plt.figure(figsize=(18, 4))
    sns.distplot(unique_rows_df['Amount'].values, color='b')

plt.show()
```
